```python
print(df.isnull().sum())
```

```python
df['Negative'].fillna(df['Negative'].mean(), inplace = True) #this command is used to find the mean for the column 'negative' and replace NaN with mean
```

```python
df.replace(r'^\s*$', np.nan, regex=True,inplace=True) #to find empty strings throughout the dataframe and replace with NaN
```

```python
df.Negative = df.Negative.astype('float64') #typecast to float
```

```python
df['Positive'].fillna(df['Positive'].mean(), inplace = True)
```

```python
df.describe() #to find out various statistical parameters for the dataframe
```

```python
corr=df.corr() #to find heatmap for correlation matrix
print(corr)
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
